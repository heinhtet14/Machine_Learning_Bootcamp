🚀 **Introduction to Object-Oriented Programming (OOP)**

Object-Oriented Programming (OOP) is a programming approach that helps you organize your code into objects. In OOP, an object bundles both data and behaviors together, similar to how an entity in the real world has attributes (like color or size) and actions (like moving or stopping).

PyTorch itself uses OOP to organize neural network components into classes and objects, making it easier to extend and modify models. Let’s explore the core concepts:

**Class**: A blueprint for creating objects.

**Object**: An instance of a class with real values.

**Inheritance**: Reusing and extending existing code.

We’ll use these concepts to build PyTorch models.

🏛️ **Class vs Instance: Understanding the Difference**

In Object-Oriented Programming (OOP), it's essential to understand the distinction between a class and an instance.

**What Is a Class?**

A class is like a blueprint or a template. It defines the properties (attributes) and behaviors (methods) that the objects created from it will have. Think of it like the blueprint of a car. It tells you how to build the car but isn’t a car itself until you use the blueprint to create one.


Using the Car class, you can create multiple instances—each representing a different car.
 Classes define functions called methods, which identify the behaviors and actions that an object created from the class can perform with its data. A class is a blueprint for how to define something. It doesn’t actually contain any data.

 **What is an Instance?**

 While the class is the blueprint, an instance is an object that’s built from a class and contains real data. It represents a realization of the blueprint. For example, if Car is a blueprint, an instance is the actual car you drive.

⭕**How Do You Inherit From Another Class in Python?**

Inheritance is the process by which one class takes on the attributes and methods of another. Newly formed classes are called child classes, and the classes that you derive child classes from are called parent classes.

Child classes can override or extend the attributes and methods of parent classes. In other words, child classes inherit all of the parent’s attributes and methods but can also specify attributes and methods that are unique to themselves.
